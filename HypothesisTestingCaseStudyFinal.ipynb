#### c. Inrest rate varies for different purpose of loans

```python
loans.head()
```

```python
loans['Loan.Purpose'].unique()
```

```python
from scipy.stats import f_oneway

interest_rates_by_purpose = []

# Iterate over each loan purpose
for purpose in loans['Loan.Purpose'].unique():
    # Select interest rates for the current purpose
    interest_rates = loans.loc[loans['Loan.Purpose'] == purpose, 'Interest.Rate']
    interest_rates_by_purpose.append(interest_rates)

# Perform ANOVA
f_statistic, p_value = f_oneway(*interest_rates_by_purpose)

print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

### BUSINESS PROBLEM-3:

Determine what effect, if any, the reengineering effort had on the
incidence behavioral problems and staff turnover. i.e To determine if the reengineering effort
changed the critical incidence rate. Isthere evidence that the critical incidence rate
improved?


```python
df = pd.read_csv("C:/Users/Kaniska Jain/Documents/DATA SCIENCE/PYTHON/case studies/4. Basic Statistics - Hypothesis Testing/Treatment_Facility.csv")
df
```

```python
df.rename(columns={'VAR4':'TRFF','VAR5':'CI'}, inplace=True)
df
```

```python
# we split the data into 2 parts pre and post reengineering

pre = df[df.Reengineer == 'Prior']
pre
```

```python
post = df[df.Reengineer == 'Post']
post
```

```python
# Checking Employee Turnover before and after
# H0: There is no change in turnover
# Ha: There is some change in the turnover
# alpha = 0.05
t_stat1, p_value1 = stats.ttest_ind(pre.Employee_Turnover,post.Employee_Turnover)
print('The t statistic is:', t_stat1)
print("The p_value is:", p_value1)

# since the p value is greater than our choosen significance level we can reject the alternate hypothesis
```

```python
# Checking the TRFF
# H0: There is no change in TRFF
# Ha: There is some change in TRFF
# alpha = 0.05
t_stat1, p_value1 = stats.ttest_ind(pre.TRFF,post.TRFF)
print('The t statistic is:', t_stat1)
print("The p_value is:", p_value1)

# since the p value is lower than our choosen significance level we can reject the null hypothesis that state
# there is no change in the TRFF after the reengineering
```

```python
# Checking the CI
# H0: There is no change in CI
# Ha: There is some change in CI
# alpha = 0.05
t_stat1, p_value1 = stats.ttest_ind(pre.CI,post.CI)
print('The t statistic is:', t_stat1)
print("The p_value is:", p_value1)

# since the p value is greater than our choosen significance level we can reject the alternate hypothesis
# that there is some change in the CI post engineering
```
